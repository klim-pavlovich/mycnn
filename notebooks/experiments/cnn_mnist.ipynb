```python
# Импортируем нужные модули
import sys
import os
sys.path.append(os.path.abspath('C:/Users/klimm/Desktop/lab2'))

import numpy as np
import matplotlib.pyplot as plt

from utils.load_mnist import load_mnist_data
from utils.split_train_validation import split_train_validation
from utils.one_hot_encode import one_hot_encode
from sklearn.model_selection import train_test_split

from src.models.le_net_5 import LeNet5
```

Загружаем MNIST и разделяем выборку

```python

# Загружаем MNIST данные
X_train, Y_train, X_test, Y_test = load_mnist_data()

# Проверяем размерности исходных данных
print(f"Исходные данные:\nX_train: {X_train.shape}, Y_train: {Y_train.shape}")
print(f"X_test: {X_test.shape}, Y_test: {Y_test.shape}")

# Разделяем обучающую выборку на 70% для обучения и 30% для валидации
X_train, X_test, Y_train, Y_test = split_train_validation(X_train, Y_train)

# Уменьшаем размер обучающей выборки до 1/10
X_train_small, _, Y_train_small, _ = train_test_split(X_train, Y_train, test_size=0.9, stratify=Y_train, random_state=42)

# Уменьшаем размер тестовой выборки до 1/10
X_test_small, _, Y_test_small, _ = train_test_split(X_test, Y_test, test_size=0.9, stratify=Y_test, random_state=42)

# Проверяем размеры новых выборок
print(f"После уменьшения:\nX_train_small: {X_train_small.shape}, Y_train_small: {Y_train_small.shape}")
print(f"X_test_small: {X_test_small.shape}, Y_test_small: {Y_test_small.shape}")
```

Нормализуем и добавляем метки

```python
# # Добавляем ось каналов
# X_train = X_train[..., np.newaxis]
# X_test = X_test[..., np.newaxis]

# # Нормализуем данные об изображениях в диапазон [0, 1]
# X_train = X_train / 255.0
# X_test = X_test / 255.0

# # Кодируем метки
# Y_train = one_hot_encode(Y_train)
# Y_test = one_hot_encode(Y_test)

# Добавляем ось каналов
X_train_small = X_train_small[..., np.newaxis]
X_test_small = X_test_small[..., np.newaxis]

# Нормализуем данные об изображениях в диапазон [0, 1]
X_train_small = X_train_small / 255.0
X_test_small = X_test_small / 255.0

# Кодируем метки
Y_train_small = one_hot_encode(Y_train_small)
Y_test_small = one_hot_encode(Y_test_small)
```

```python
print(f"X_train_small min: {X_train_small.min()}, max: {X_train_small.max()}")
print(f"Y_train_small min: {Y_train_small.min()}, max: {Y_test_small.max()}")
```

```python
# Инициализация модели
model = LeNet5()

# Обучение модели
epochs = 5
learning_rate = 0.0001
batch_size = 32
model.train(X_train_small, Y_train_small, epochs, learning_rate, batch_size, X_test_small, Y_test_small)
```
